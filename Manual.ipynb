## Interpolation of Velocities for Lagrangian Particles

`interpolation` launches the kernel that interpolates velocities for each particle according using a bilinear interpolation:

\begin{align}
u_{p} = &&(1-dx) \, (1-dy) \, (1-dz) \, u_{A} +\\
        &&   dx \, (1-dy) \, (1-dz) \, u_{B} + \\
        &&   dx \, dy \, (1-dz) \, u_{C} + \\
        &&   (1-dx) \, dy \, (1-dz) \, u_{D} + \\
        &&   (1-dx) \, (1-dy) \, dz \, u_{E} + \\
        &&   dx \, (1-dy) \, dz \, u_{F} + \\
        &&   dx \, dy \, dz \, u_{G} + \\
        &&   (1-dx) \, dy \, dz \, u_{H}\\
\end{align}

`interpolation2` launches the kernel that interpolates velocities for each particle according taking a normal average:
$$u_{p} = \frac{u_{A} + u_{B} + u_{C} + u_{D} + u_{E} + u_{F} + u_{G} + u_{H}}{8}$$

```python
#Getting interpolation function(interpolation2: bilinear, interpolation: averages)
interpolate = mod.get_function('interpolation2')
```

```python
#Initializing Velocities to Zero
partVelu = np.zeros(numParticles, dtype=np.float32)
partVelv = np.zeros(numParticles, dtype=np.float32)
partVelw = np.zeros(numParticles, dtype=np.float32)
```

```python
#Interpolation
interpolate(pycuda.driver.In(oldPosX),
     pycuda.driver.In(oldPosY),
     pycuda.driver.In(oldPosZ),
     pycuda.driver.In(xi),
     pycuda.driver.In(yi),
     pycuda.driver.In(zi),
     pycuda.driver.In(u_1d),
     pycuda.driver.In(v_1d),
     pycuda.driver.In(w_1d),
     pycuda.driver.Out(partVelu),
     pycuda.driver.Out(partVelv),
     pycuda.driver.Out(partVelw),
     np.float32(Sim.particlesGPU.local_ximin),
     np.float32(Sim.particlesGPU.local_ximax),
     np.float32(Sim.particlesGPU.local_yimin),
     np.float32(Sim.particlesGPU.local_yimax), 
     np.float32(Sim.particlesGPU.local_zimin),
     np.float32(Sim.particlesGPU.local_zimax), 
     np.float32(nx), np.float32(ny), np.float32(nz), 
     np.int32(gw),
     np.float32(dx), np.float32(dy), np.float32(dz), 
     np.int32(numParticles),
     grid=(blocks,1), block=(threads_per_block,1,1))
```

## Integration of Positions for Lagrangian Particles

Each thread computes a forward Euler integration for the particles using interpolated velocities. 
$$
x_{i,n+1} = x_{i,n} + v_{i}(x_{n},t) \, dt
$$

```python
#Getting compiled CUDA function for integration
#'integrateBomex'       : Bomex boundary conditions (Periodic in X and Y and reflection in Z)
#'integratePeriodic'    : Periodic boundary conditions in X, Y, and Z
#'test_integrateSimple' : Reflections in X, Y, and Z
integrate1 = mod.get_function('integrateBomex')
#Copying values to GPU every iteration
newPosX = np.zeros(numParticles, dtype=np.float32)
newPosY = np.zeros(numParticles, dtype=np.float32)
newPosZ = np.zeros(numParticles, dtype=np.float32)
```

```python
integrate1(pycuda.driver.In(oldPosX),
           pycuda.driver.In(oldPosY),
           pycuda.driver.In(oldPosZ),
           pycuda.driver.In(partVelu),
           pycuda.driver.In(partVelv),
           pycuda.driver.In(partVelw),
           pycuda.driver.Out(newPosX),
           pycuda.driver.Out(newPosY),
           pycuda.driver.Out(newPosZ),
           np.float32(dt),
           np.float32(Sim.particlesGPU.local_ximin),
           np.float32(Sim.particlesGPU.local_ximax),
           np.float32(Sim.particlesGPU.local_yimin),
           np.float32(Sim.particlesGPU.local_yimax),
           np.float32(Sim.particlesGPU.local_zimin),
           np.float32(Sim.particlesGPU.local_zimax),
           np.int32(numParticles),
           grid=(blocks,1), block=(threads_per_block,1,1))
```

```python
#Uncomment this line to deactivate inline plotting and create an interactive 3D plot 
#%matplotlib 
#Plotting integrated positions
fig3 = plt.figure(figsize=(16,9))
ax1 = fig3.add_subplot(121, projection='3d')
ax1.plot(oldPosX, oldPosY, oldPosZ,'o', markersize=7, color='red', alpha=0.5)
ax1.plot(newPosX, newPosY, newPosZ,'o', markersize=7, color='red', alpha=1)
for xp, yp, zp, xp_n, yp_n, zp_n in zip(oldPosX, oldPosY, oldPosZ, newPosX, newPosY, newPosZ):
    vPart= Arrow3D([xp,xp_n], [yp,yp_n],[zp,zp_n], mutation_scale=10, lw=2, arrowstyle="-|>", color="red")
    ax1.add_artist(vPart)
plt.title('Integrated Positions')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')
ax1.set_xlim([ximin,ximax])
ax1.set_ylim([yimin,yimax])
ax1.set_zlim([zimin,zimax])

ax2 = fig3.add_subplot(122, projection='3d')
ax2.plot(oldPosX, oldPosY, oldPosZ,'o', markersize=7, color='red', alpha=0.5)
#ax2.plot(newPosX, newPosY, newPosZ,'o', markersize=7, color='red', alpha=1)
for xp, yp, zp, up, vp, wp in zip(oldPosX, oldPosY, oldPosZ, partVelu, partVelv, partVelw):
    vPart= Arrow3D([xp,(xp+dt*up)], [yp,(yp+dt*vp)],[zp,(zp+dt*wp)], mutation_scale=10, lw=2, arrowstyle="-|>", color="black")
    ax.add_artist(vPart)
plt.title('Integrated Positions')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')
ax1.set_xlim([ximin,ximax])
ax1.set_ylim([yimin,yimax])
ax1.set_zlim([zimin,zimax])
```
